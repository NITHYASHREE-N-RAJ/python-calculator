```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
df=pd.read_csv("/content/Mall_Customers.csv")
```

```python
x=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
agg_model=AgglomerativeClustering(n_clusters=200)
```

```python
df['cluster']=agg_model.fit_predict(x)
```

```python
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'],c=df['cluster'],cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Agglomerative Clustering')
plt.show()
```

```python
df.to_csv('clustered_data.csv',index=False)
print("Agglometerative clustred dataset saved as agglo_clustered_data.csv")
```
